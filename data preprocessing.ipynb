```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Dataset.csv")
data_df = pd.DataFrame(data)
data_df["Date"] = pd.to_datetime(data_df["Date"])
data_df.set_index("Date", inplace = True)
```

```python
data_df = data_df.astype({"BTC" : 'int64', "NYSE": "int64", "NASDAQ": "int64", "LSE": "int64", "LSE_Volume": "int64"}) 
```
